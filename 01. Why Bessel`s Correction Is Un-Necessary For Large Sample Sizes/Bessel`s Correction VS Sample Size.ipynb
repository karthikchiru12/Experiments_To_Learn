**At glance:**

- Imagine you have 10 coins in your pocket. And the average of coins is 0.5
- To know all the coins in the pocket, how many coins you need to know ?
- Just 9!. because already we know the average (0.5).
- The 10nth coin is dependent on rest of the 9 coins, whereas the 9 coins can be anything(free to vary).
- So in the above situation the number of things that can be varied are 9 (degrees of freedom)

**Importing the necessary packages**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Simple script to calculate and plot the biased v/s unbiased variance**

```python
class Compare_Biased_And_Unbiased_Variance:
    
    def __init__(self,s_lower,s_upper,distribution="Gaussian"):
        
        self.s_upper = s_upper
        self.s_lower = s_lower
        self.distribution = distribution
        self.biased_var  = []
        self.unbiased_var = []
        self.n = []
    
    def plot(self):
        
        for i in range(self.s_lower,self.s_upper):
            data = np.random.randn(i)
            mean_centered_data = data - np.mean(data)
            self.unbiased_var.append( sum(mean_centered_data**2) / (i-1) )
            self.biased_var.append( sum(mean_centered_data**2) / (i) )
            self.n.append(i)
        plt.figure(figsize=(20,10))
        plt.plot(self.n,self.biased_var,'r',marker='o',label="Biased variance")
        plt.plot(self.n,self.unbiased_var,'b',marker='s',label="Unbiased variance")
        plt.xlabel("Sample size")
        plt.ylabel("Variance")
        plt.legend()
        plt.xlim([self.s_lower, self.s_upper])
        plt.xticks(self.n)
        plt.title(self.distribution+" distributed data \n", fontsize=24)
        plt.show()      
```

**When sample size is in range of 2-50:**

```python
Compare_Biased_And_Unbiased_Variance(2,50).plot()
```

**When sample size is in range of 50-100:**

```python
Compare_Biased_And_Unbiased_Variance(50,100).plot()
```

**When sample size is in range of 100-150:**

```python
Compare_Biased_And_Unbiased_Variance(100,150).plot()
```

**When sample size is in range of 800-850:**

```python
Compare_Biased_And_Unbiased_Variance(800,850).plot()
```

**Conclusions:**

- When the sample size is < 15, the differrence between biased variance and unbiased variance is high.
- When the sample size increases the difference becomes almost insignificant
